```python
import random

def palabra_random(lista):
    return random.choice(lista)

def dificultad():
    lst_palabras5 = ['calle','coche','numero','papel','chino','carta','color','negro','curso','datos','flama','zebra']
    lst_palabras10 = ['microbiano','mobiliario','nebulizado','noctambulo','utensilios','urbanistas','yuxtaponer','yugoeslavo','vulnerable','viajaremos','vegetacion','vacaciones','vehementes','tabacalera','tecnocratas','tecnologia']
    lst_palabras15 = ['oficializaremos','originariamente','ornamentaciones','glorificaciones','gravitacionales','pluripartidista','psicopatologias']

    dif = int(input("¿Qué grado de dificultad? Elige un número: \n"
                          "1. facil \n"
                          "2. medio \n"
                          "3. dificil \n"))
    global pala_hang
    while dif <= 4:
        if dif == 1:
            pala_hang = palabra_random(lst_palabras5)
            return pala_hang
        elif dif == 2:
            pala_hang = palabra_random(lst_palabras10)
            return pala_hang
        elif dif == 3:
            pala_hang = palabra_random(lst_palabras15)
            return pala_hang
        else:
            dif = int(input('Selecciona una dificultad: 1, 2 o 3: '))
def jugar_otravez():
    resp = input("¿Jugar otra vez?: si o no: ")
    if resp == 'si':
        print("Juguemos")
        dificultad()
    else:
        global jugar
        jugar = "no"
        print("Gracias por jugar")


def ahorcado():
    global jugar
    jugar = input('¿Listo para jugar? si o no: ').lower()

    while jugar == 'si':
        dificultad()
        letras_adivinadas = []
        guiones = []
        oportunidades = 5

        while oportunidades > 0 and "".join(guiones) != pala_hang:

            letra = input("Tu suposición: ").lower()
            guiones = []

            errores = [1 for char in letra if char not in pala_hang]
            oportunidades -= sum(errores)

            for idx,let in enumerate(pala_hang):
                if let in letra:
                    letras_adivinadas.append(let)
                    guiones.append(let)
                elif let in letras_adivinadas:
                    guiones.append(let)

                else:
                    guiones.append("_ ")

            print("\n Te quedan {} oportunidades".format(oportunidades)
                  + "\n Tu palabra: ".format(guiones)
                  + "\n Letra adivinada: " + ", ".join(letras_adivinadas))

            print(''.join(guiones))

        if "".join(guiones) == pala_hang:
            print("\n\nFELICIDADES,GANASTE!!\n")
            jugar_otravez()
        else:
            print("\nLo sentimos , el juego ha terminado!\nTu palabra era: {}".format(pala_hang))
            jugar_otravez()
    else:
        print("Vuelve pronto")
```
